Pandas and Data Visualization: A Powerful Combination
Pandas is a cornerstone library in Python for data manipulation and analysis. While not primarily a visualization library, it offers built-in plotting functionalities that are often sufficient for exploratory data analysis (EDA).

Pandas Plotting Basics
Pandas leverages Matplotlib as its backend for plotting. The primary method for creating plots is the plot() method,

Q. Creation of a dataframe with 20 rows and 10 coloumns. The dataframe has been created with random values from range 1 - 100.

```python
import matplotlib.pyplot as plt
```

```python
plot = [1, 5, 8, 9, 2, 0, 3, 10, 4, 7]
plot1 =[ 3,8,9,2,1,2,4,7,6,6]
```

```python
plt.plot(range(1,11))
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Simple Plot')
```

```python
plt.plot(plot, plot1)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Simple Plot')
plt.show()
```

```python
ax = plt.axes()
ax.grid()
```

Pie Chart of Sum of Colummns and their ratio.

```python
newlist = [5,8,9,10,4,7]
colors = ['darkblue', 'green', 'red', 'cyan', 'magenta', 'yellow']
lables = ['a','b','c','d','e','f']
explod = [0,0.2,0,0,0,0]
plt.pie(newlist, labels=lables, colors=colors, autopct='%1.1f%%', startangle=140, explode=explod)
```

Box Plot of all coloumns.

```python
newlist1 = [5,8,9,10,4,7]
colors = ['darkblue', 'green', 'red', 'cyan', 'magenta', 'yellow']
lables = ['a','b','c','d','e','f']
width = [0.7,0.8,0.7,0.7,0.7,0.7]

plt.bar(lables, newlist1, color=colors, width=width)
```
