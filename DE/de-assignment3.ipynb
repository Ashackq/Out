# Interquartile Range (IQR)
The Interquartile Range (IQR) is a measure of statistical dispersion, which indicates the spread of the middle 50% of your data.

# Why is it useful?
* Robust to outliers: Unlike the range (which is sensitive to extreme values), the IQR focuses on the middle portion of the data, making it less affected by outliers.

   
* Complements the median: When dealing with skewed data, the median is often a better measure of central tendency than the mean. The IQR complements the median by providing information about the spread of the data.   


```python
import pandas as pd
import numpy as np
```

# Steps To Perform IQR

* Import numpy: Imports the NumPy library for efficient numerical operations.
* Define the function: Defines a function named calculate_iqr that takes a dataset as input.
* Input validation: Checks if the input data is a NumPy array or a list. Raises a TypeError if not.
* Convert to NumPy array: Converts the input data to a NumPy array for efficient calculations.
* Calculate quartiles: Uses np.percentile to calculate the first quartile (Q1, 25th percentile) and third quartile (Q3, 75th percentile) of the data.
* Calculate IQR: Calculates the IQR as the difference between Q3 and Q1.
* Return IQR: Returns the calculated IQR.

```python
def iqrc(l1):
    if(type(l1) == list):
        l1.sort()
    elif (type(l1)==pd.DataFrame):
        l1 = l1.sort_values(by="number")
    data = np.array(l1["number"])
    q1,q3 = np.percentile(data,25),np.percentile(data,75)
    iqr = q3 - q1
    ub , lb = q3+(1.5*iqr),q1-(1.5*iqr)
    print(f"Iqr:{iqr}\nUpper Bound:{ub}\nLower Bound:{lb}")
```

```python
dataset= [11,10,12,14,12,15,14,13,15,102,12,14,17,19,107, 10,13,12,14,12,108,12,11,14,13,15,10,15,12,10,14,13,15,10]
```

```python
iqrc(dataset)
```
