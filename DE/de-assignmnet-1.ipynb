# Data Collection
Data collection is the systematic process of gathering and recording information on specific variables or topics.

# Common data handling operations with Pandas and NumPy:

* Loading data: Reading data from various file formats (CSV, Excel, JSON, etc.) into Pandas DataFrames.
* Data cleaning: Handling missing values, outliers, inconsistencies, and duplicates.
* Data transformation: Creating new features, reshaping data, and aggregating data.
* Data exploration: Summarizing data, visualizing distributions, and identifying patterns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Functions Used for Data Collection
* pd.read_csv('iris.csv'): Reads a CSV file into a Pandas DataFrame.
* head(), tail(): Display the first/last few rows of the DataFrame.
* shape: Returns the number of rows and columns.
* describe(): Generates descriptive statistics for numerical columns.
* info(): Provides information about the DataFrame, including column names, data types, and non-null counts.
* duplicated().sum(): Counts the number of duplicate rows.
* drop_duplicates(): Removes duplicate rows.
* isnull().sum(): Counts the number of missing values in each column.
* fillna(): Fills missing values with specified values (here, the mean).

```python
# the dataset 'iris.csv' 
iris_data = pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
iris_data
```

```python
# Display the first few rows
print(iris_data.head())

# Display the last few rows
print(iris_data.tail())

# Get the shape of the dataset (rows, columns)
print(iris_data.shape)

# Get summary statistics
print(iris_data.describe())

# Get information about the dataset
print(iris_data.info())
```

```python
# Check for duplicates
print(iris_data.duplicated().sum())

# Remove duplicates (if necessary)
iris_data = iris_data.drop_duplicates()

# Check for missing values
print(iris_data.isnull().sum())

# Handling missing values (example: filling with mean)
iris_data.fillna(iris_data["SepalWidthCm"].mean(), inplace=True)
```

```python
print("Cleaning Stage- 1")
print(iris_data)
```

# Vistualise and Understand the Data

```python
sns.histplot(iris_data['SepalLengthCm'], kde=True)
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=iris_data)
plt.show()
```
