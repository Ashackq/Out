```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1, 101, size=(20, 10)), columns=[f'Col_{i}' for i in range(1, 11)])

print(df)
```

Creation of a dataframe with 20 rows and 10 coloumns. The dataframe has been created with random values from range 1 - 100.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
df['Col_1'].plot(kind='bar')
plt.title('Bar Chart')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

Bar Chart representation of Col_1 Values.

```python
plt.figure(figsize=(10, 6))
df['Col_2'].plot(kind='line')
plt.title('Line Chart ')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

Line Chart representation of Col_2 Values.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Col_3'], df['Col_4'])
plt.title('Scatter Plot between Col_3 and Col_4')
plt.xlabel('Col_3')
plt.ylabel('Col_4')
plt.show()
```

Scatter Plot of Col_3 and Col_4 values.

```python
plt.figure(figsize=(10, 6))
df['Col_5'].plot(kind='hist', bins=10)
plt.title('Histogram of Col_5')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Histogram of Col_5

```python
plt.figure(figsize=(10, 6))
df.sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Sum of All Columns')
plt.ylabel('')
plt.show()
```

Pie Chart of Sum of Colummns and their ratio.

```python
plt.figure(figsize=(12, 8))
df.plot(kind='box')
plt.title('Box Plot of All Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

Box Plot of all coloumns.
